```python
from corpusanalyzer import CorpusAnalyzer
import nltk
from nltk.corpus.reader import PlaintextCorpusReader
```

```python
data_dir = './data/'
chilit = PlaintextCorpusReader(root = data_dir + "ChiLit", fileids = r'.+\.txt')
```

```python
# Initialize analyzer
analyzer = CorpusAnalyzer(chilit)

# Perform analysis
individual_stats, corpus_summary, df = analyzer.analyze_corpus()
```

```python
 # Print summary
analyzer.print_summary(individual_stats, corpus_summary)

# The DataFrame 'df' contains all the statistics and can be used for further analysis
print(f"\nDataFrame shape: {df.shape}")
print("\nDataFrame columns:", list(df.columns))
```

```python
# Additional analysis
print(f"\nCorrelation between document length and lexical density:")
print(f"  Correlation coefficient: {df['total_words'].corr(df['lexical_density']):.3f}")
```

```python
# Analysis of Document Lenghts
doc_lengths, stats = analyzer.analyze_document_lengths(individual_stats)
analyzer.plot_length_distribution(doc_lengths)
```
